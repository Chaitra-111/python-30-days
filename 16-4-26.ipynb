#### Task 1

```python
marks = []

for i in range(10):
    m = int(input(f"Student {i+1}: "))
    marks.append(m)

print("Marks List:", marks) 
```

#### Task 2

```python
average = sum(marks) / len(marks)
highest = max(marks)
lowest = min(marks)
print("Average: ", average)
print("Highest: ", highest)
print("Lowest: ", lowest)
```

#### Task 3

```python
count = 0
for m in marks:
    if m >= 85:
        grade = 'A'
    elif m >= 70:
        grade = 'B'
    elif m >= 50:
        grade = 'C'
    else:
        grade = 'F'
    print(f"Marks: {m} Grade: {grade}")

print('---------Above 75---------')
for m in marks:   
    if m > 75:
        count += 1
print("Students above 75:", count)
```
