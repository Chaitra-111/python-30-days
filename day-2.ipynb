## 3. Convert to Uppercase and Lowercase

```python
a=input("enter the string: ")
print("the uppercase of string is : ",a.upper())
print("the lower case of string is : ",a.lower())
```

## 4. Reverse a String

```python
#indexing
a=input("enter the string: ")
print(a[::-1])
#user input
b=input("enter the string: ")
rev=""
for i in b:
    rev=i+rev
print(rev)
```

## 9. Check if a Substring Exists

```python
a=input("enter the string: ")
word=input("enter the search substring")
if word in a:
    print(f"{word} is in the string")
else:
    print(f"{word} is not in the string")
```

## 10. Count Occurrences of a Character

```python
word = "Chaitra"
D = {}
for i in word:
    if i in D:
        D[i]= D[i]+1
    else:
        D[i] = 1
print(D)
```

## 15. Check Start and End of a String

```python
a=input("enter the string: ")
print("the string starts with :", a.startswith("python"))
print("the string ends with : ", a.endswith("data"))
print("the string starts with : :", a.startswith("teach"))
```

## 16. Find Position of a Substring

```python
a = input("Enter the string: ").split() 
word = input("Enter the search word: ")
if word in a:
    print(f"'{word}' is in the string at the word position {a.index(word)}")
else:
    print(f"'{word}' is not in the string")
```

## 18. Capitalize and Swap Case

```python
a=input("enter the string: ")
cap=a.capitalize()
swap=a.swapcase()
print("capitalize: ",cap)
print("swap: ",swap)
```

## 20. Print Characters with Index

## 22. Check if String is Empty

```python
#using length
a=input("enter the string: ")
if len(a)==0:
    print("string is empty")
else:
    print("string is not empty")
    
#using empty string
a=input("enter the string: ")
if a=='':
    print("string is empty")
else:
    print("string is not empty")
```
